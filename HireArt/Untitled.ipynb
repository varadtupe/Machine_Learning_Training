```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv',sep='\t')
```

```python
data['Date of Contact'] = pd.to_datetime(data['Date of Contact'])
```

```python
grpByDate = pd.DataFrame({'count':data.groupby('Date of Contact')['Date of Contact'].count()})
```

```python
grpByDate = grpByDate.reset_index()
```

```python
grpByDate.plot(x=grpByDate['Date of Contact'])
```
